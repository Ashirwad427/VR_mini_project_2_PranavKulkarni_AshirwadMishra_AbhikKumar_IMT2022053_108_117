# Data Curation Strategy 1

## Description :
### Rather than just sampling the data on the basis of counts, we shall also take into account the metadata of the products. The reason is to ensure a diverse dataset. For example, if the data has a overwhelming number of products of a particular colour, then our data will carry forward that bias. Using the metadata will help curb that bias.

### Steps:
- ### Assign a score to the data. If the data has a color, then it gets one point, if it has item_keywords it gets 1 point and if it has bullet_points then it gets 1 point.
- ### Filtering out the low scoring ones.
- ### Selecting diverse samples for each product type

## Assigning a score

```python
type(metadata_df.loc[1])
```

```python
def assign_score(row : pd.core.series.Series) -> int:
    """
    This function takes in a row of the dataset and assigns a score.
    color         : 1
    item_keywords : 1
    bullet_points : 1
    @params :
        A row of the dataframe
    @return :
        Score for the row of the dataset
    """

    score = 0
    if pd.notna(row['color']) and len(row['color'].strip()) > 0:
        score += 1

    if pd.notna(row['bullet_point']) and len(row['bullet_point'].strip().split(",")) > 3:
        score += 1

    if pd.notna(row['item_keywords']) and len(row['item_keywords'].strip().split(",")) > 3:
        score += 1

    return score
```

```python
metadata_df['qa_score'] = metadata_df.apply(assign_score, axis = 1)
```

```python
metadata_df
```

## Filtering out the low scoring ones

```python
metadata_df['qa_score'].value_counts()
```

```python
filtered_df = metadata_df[metadata_df['qa_score'] >= 2]
```

```python
filtered_df
```

## Selecting Diverse Samples for Each Product Type

```python
sampled_rows = []
```

```python
max_samples_per_class = 200
```

```python
for cat, group in filtered_df.groupby('product_type'):

    group_sorted = group.sort_values('qa_score', ascending=False)

    group_dedup = group_sorted.drop_duplicates(subset=['color', 'item_keywords', 'bullet_point'])

    top_k = group_dedup.head(max_samples_per_class)

    sampled_rows.append(top_k)
```

```python
len(sampled_rows)
```

```python
final_sample_df = pd.concat(sampled_rows).drop_duplicates(subset=['main_image_id'])
```

```python
final_sample_df
```

```python
final_sample_df = final_sample_df.sample(frac=1, random_state=42).reset_index(drop=True)
```
